```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
add = 'D:\data\data\stage4.xlsx'
```

```python
data1 =  pd.read_excel(add,sheet_name = 0,index_col=0)
data1
```

```python
dataa = data1['投资时间']
dataa = set(dataa)
dataa = list(dataa)
dataa.sort()
dataa
```

```python
data1.set_index('投资时间',drop=True, append=False, inplace=True)
```

```python
ans = 0
df = pd.DataFrame()
for i in dataa:
    x = data1.loc[i]
    if isinstance(x,pd.Series)== True:
            continue
    if ans == 0:
        x.set_index('企业名称',drop=False, append=False, inplace=True)
    else :
        ans=ans+1
    
    y = x['企业名称']
    y = set(y)
    y = list(y)
    y.sort()

    
    for j in y:
        z = x.loc[j]
        
        if isinstance(z,pd.Series)== True:
            continue
        else:
            zz = z['机构名称']
            zz = set(zz)
            zz = list(zz)
            zz.sort()
#            print(zz)
#            print('\n')
            
            for i1 in range(len(zz)):
                for i2 in range(i1+1,len(zz)):
                    x1 = zz[i1]
                    x2 = zz[i2]
                    c={'投资次数':1,
                       '机构名称1':x1,
                       '机构名称2':x2}
                    
                    df1 = pd.DataFrame(c,index=[0])
                    df = df.append(df1, ignore_index=True)
```
