```python
data_all =  pd.read_excel('D:\data\data\investments_clear3(1).xlsx',sheet_name = 0,index_col=0)
data_all
```

```python
qy = [1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]
```

```python
data_all.set_index('投资时间',drop=True, append=False, inplace=True)
```

```python
data_all[0:40]
print(data_all.columns)
```

```python
ans = 0
cnt = 0
res = []
for i in qy:
    x = data_all.loc[i]
    if isinstance(x,pd.Series) == True:
        continue
    if ans == 0:
        x.set_index('一级字母代码',drop=False, append=False, inplace=True)
    else :
        ans=ans+1 
    
    first = x['一级字母代码']
    first = set(first)
    first = list(first)
    
    for j in first:
        y = x.loc[j]
        if isinstance(y,pd.Series) == True:
            continue
        if cnt == 0:
            y.set_index('二级代码',drop=False, append=False, inplace=True)
        else :
            cnt=cnt+1 
        sec = y['二级代码']
        sec = set(sec)
        sec = list(sec)
        
        
        for k in sec:
            z = y.loc[k]
            if isinstance(z,pd.Series) == True:
                continue
                
            zz = z['机构名称']
            zz = set(zz)
            zz = list(zz)
            if len(zz) == 1 :
                continue
            res.append(zz)
```
