```python
dataa = data1['投资时间']
dataa = set(dataa)
dataa = list(dataa)
dataa.sort()
dataa
```

```python
shanchu = data1['投资时间'].value_counts()
shanchu
```

```python
for i in dataa:
    nih = shanchu[i]
    if nih == 1:
        dataa.remove(i)
len(dataa)#得2330#
```

```python
data1.set_index('投资时间',drop=True, append=False, inplace=True)
```

```python
data1.loc['2015-12-29']
```

```python
ans = 0
df = pd.DataFrame()
for i in dataa:
    x = data1.loc[i]
    if ans == 0:
        x.set_index('企业名称',drop=False, append=False, inplace=True)
    else :
        ans=ans+1

    
    y = x['企业名称']
    y = set(y)
    y = list(y)
    y.sort()


    for j in y:
        z = x.loc[j]
        
        if isinstance(z,pd.Series)== True:
            continue
        else:
            zz = z['机构名称']
            zz = set(zz)
            zz = list(zz)
            zz.sort()
#            print(zz)
#            print('\n')
            
            for i1 in range(len(zz)):
                for i2 in range(i1+1,len(zz)):
                    x1 = zz[i1]
                    x2 = zz[i2]
                    c={'投资次数':1,
                       '机构名称1':x1,
                       '机构名称2':x2}
                    
                    df1 = pd.DataFrame(c,index=[0])
                    df = df.append(df1, ignore_index=True)
                    
                    
                        
print(df)
```
